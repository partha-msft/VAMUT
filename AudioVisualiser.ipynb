```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(15,4))
audFilePath = 'C:\\Users\\pasm\\Documents\\Hack2020\\Silencio\\Inputs\\SpeechWithMinNoise.wav'
data,sr = librosa.load(audFilePath)
data,_ = librosa.effects.trim(data)
librosa.display.waveplot(data,sr=sr)
```

```python
n_fft = 2048
hop_length = 512
n_mels = 128

S = librosa.feature.melspectrogram(data, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
S_DB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
```
